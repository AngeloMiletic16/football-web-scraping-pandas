## Reading .csv from Multiple URLs with Pandas

##### https://www.football-data.co.uk/mmz4281/2526/E0.csv

##### link: root + season + league

```python
# a root variable - all links has this root in common:
root = "https://www.football-data.co.uk/mmz4281/"
```

```python
#list of leagues
leagues = ["E0","E1","E2","E3","EC"]
frames = []

#LOOPING THORUGH LEAGUES, READING MULTIPLE CSV AND APPEND IT INTO A LIST
for league in leagues:
    df = pd.read_csv(root + "2526" + "/" + league + ".csv")
    frames.append(df)
```

```python
#length of frames
len(frames)
```

```python
#sbow 1st, 2nd and 3rd elemtn
frames[4]
```

## Multiple Seasons 

```python
#list of leagues
leagues = ["E0","E1","E2"]
frames = []

#LOOPING THORUGH LEAGUES, READING MULTIPLE CSV AND APPEND IT INTO A LIST
for league in leagues:
    for season in range(19, 25):
        df = pd.read_csv(root + str(season) + str(season+1) + "/" + league + ".csv", encoding="unicode_escape")
        df.insert(1, "season", season)
        frames.append(df)    
```

```python
#total frames : 6 seasons x 3 leagues
len(frames)
```

```python
#show 1st and last elemnt (new season column was creates)
frames[17]
```

## Organize All the data in a Dictionary

```python
dict_countries = {
    "Spanish La Liga":"SP1", "Spanish Segunda Division":"SP2",
    "German Bundesliga":"D1",
    "English Premmier League":"E0", "English League 1":"E2", "English League 2":"E3"
}
```

```python
dict_countries["Spanish La liga"]
```

```python
dict_historical_data = {}


#list of leagues
#leagues = ["E0","E1","E2"]

#LOOPING THORUGH LEAGUES, READING MULTIPLE CSV AND APPEND IT INTO A LIST
for league in dict_countries:
    frames = []
    for season in range(19, 25):
        df = pd.read_csv(root + str(season) + str(season+1) + "/" + dict_countries[league] + ".csv", encoding="unicode_escape")
        df.insert(1, "season", season)
        frames.append(df)   
    df_concat = pd.concat(frames)
    dict_historical_data[league] = df_concat 
```
